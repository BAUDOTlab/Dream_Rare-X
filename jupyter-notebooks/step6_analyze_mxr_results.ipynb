# Results interpretation

## Summary of the previous steps
We propose to address the problem of identifying new symptoms associated with rare diseases through a biological network analysis using Random Walk with Restart (RWR).

Biological networks offer the advantage of representing various types of interactions between biological entities.

We constructed a heterogeneous network consisting of two layers:
* The first layer contains data from the challenge, including rare diseases, patients within the cohort, and symptoms associated with each individual. By navigating this layer starting from the diseases, we can bypass the need for frequency calculations, as the Random Walker (RW) will propagate towards symptoms via patients presenting those symptoms.
* The second layer contains information from the Orphanet and Human Phenotype Ontology databases, with nodes representing rare diseases and HPO terms. Edges connecting diseases exist if they share at least one common causative gene, allowing the Random Walker to propagate from one disease to another in the network. The RW can also spread to HPO terms via Disease-HPO edges, weighted based on the reported frequency of the phenotype for each disease. Through the HPO ontology, the RW can further navigate across the ontology.
* The RW can switch between the two layers using the available mappings between RareX diseases and Orphanet diseases.

The main idea is to perform RW from each RareX disease, which allows us to score symptom nodes and HPO nodes while incorporating heterogeneous associations between entities in the network.  By comparing the scores of symptom nodes and HPO nodes, we can pinpoint symptoms that have a high score but do not receive a similarly high score for their HPO equivalent according to the RW.

```python
import pandas as pd
```

```python
seedsfile = '../Diseases_names_and_seeds_numbering.tsv'
resultsdir = '../multixrank_RARE_X_diseases/results_output_DiseaseDisease_PhenotypeOntology_Weighted/'
topresults = 15
```

First, we extract seed numbers and their corresponding disease names.

```python
seeds = pd.read_csv(seedsfile, header=0, sep='\t')
seeds.head(5)
```

Then, we read each result files and store them into a dictionnary.

```python
dico_results = dict()

for index, row in seeds.iterrows():
    seed = row[2]
    disease = row[0]
    resultfile = f'{resultsdir}/results_disease_{seed}.tsv'
    dico_results[disease] = pd.read_csv(resultfile, header=0, sep='\t')
    del dico_results[disease]['empty']
```

## Disease with orphanet correspondance

There are 15 diseases with Orphanet correspondance.

```python
seeds_withOrpha = seeds[seeds['ORPHANET'] != "None"]
seeds_withOrpha
```

We load multiXrank results for each disease (used as seed) and store them into a dictionnary ``dico_results_withOrpha``.

```python
dico_results_withOrpha = dict()

for index, row in seeds_withOrpha.iterrows():
    seed = row[2]
    disease = row[0]
    resultfile = f'{resultsdir}/results_disease_{seed}.tsv'
    dico_results_withOrpha[disease] = pd.read_csv(resultfile, header=0, sep='\t')
    del dico_results_withOrpha[disease]['empty']
```
